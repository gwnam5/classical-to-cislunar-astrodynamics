```python
from pathlib import Path
import sys

# 현재 위치가 notebooks 폴더면, 한 단계 위를 프로젝트 루트로 잡음
PROJECT_ROOT = Path.cwd()

if PROJECT_ROOT.name == "notebooks":
    PROJECT_ROOT = PROJECT_ROOT.parent

if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

print("Project root:", PROJECT_ROOT)

import numpy as np

from src.constants import MU_EARTH
from src.two_body import (
    two_body_acceleration,
    specific_angular_momentum,
    specific_energy,
    eccentricity_vector,
    eccentricity,
    semi_major_axis_from_energy
)
```

```python
r = np.array([7000.0, 0.0, 0.0])      # position vector [km]
v = np.array([0.0, 7.5, 1.0])         # velocity vector [km/s]

print("r =", r, "km")
print("v =", v, "km/s")
print("mu_earth =", MU_EARTH, "km^3/s^2")
```

```python
a = two_body_acceleration(r, MU_EARTH)
h = specific_angular_momentum(r, v)
eps = specific_energy(r, v, MU_EARTH)

print("acceleration =", a, "km/s^2")
print("specific angular momentum =", h, "km^2/s")
print("specific energy =", eps, "km^2/s^2")
```

```python
print("type(r) =", type(r))
print("type(v) =", type(v))
print("type(MU_EARTH) =", type(MU_EARTH))
print("type(a) =", type(a))
print("type(h) =", type(h))
print("type(eps) =", type(eps))
```

```python
r_norm = np.linalg.norm(r)
v_norm = np.linalg.norm(v)
a_norm = np.linalg.norm(a)
h_norm = np.linalg.norm(h)

print(f"|r| = {r_norm:.6f} km")
print(f"|v| = {v_norm:.6f} km/s")
print(f"|a| = {a_norm:.9f} km/s^2")
print(f"|h| = {h_norm:.6f} km^2/s")
print(f"epsilon = {eps:.6f} km^2/s^2")
```

```python
e_vec = eccentricity_vector(r, v, MU_EARTH)
e = eccentricity(r, v, MU_EARTH)
a_sma = semi_major_axis_from_energy(r, v, MU_EARTH)

print("eccentricity vector =", e_vec)
print("eccentricity =", e)
print("semi-major axis =", a_sma, "km")
```
